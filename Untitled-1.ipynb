```python
from datetime import datetime

dataList = []
content = "EIASC Leadership Development Workshop organized by Ethiopian Islamic Affairs Supreme council (EIASC) in collaboration with Impact Makers Foundation (IMF)"
date = datetime.now().strftime("%Y-%m-%d")
for name in nameList:
    dataList.append(f"{name};{content};{date}")
```

```python
import csv

# specify the file name
filename = "output.csv"

# writing to csv file
with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    # writing the data rows
    csvwriter.writerow(["name", "content", "date"])
    for data in dataList:
        csvwriter.writerow(data.split(';'))
```
